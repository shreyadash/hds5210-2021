## Part 1: CHA2DS2-VASc

This scoring mechanism for Atrial Fibrillation Stroke uses 7 inputs:
* Age (Number)
* Sex (Male / Female)
* CHF History (True / False)
* Hypertension History (True / False)
* Stroke History (True / False)
* Vascular Disease History (True / False)
* Diabetes History (True / False)

Fill out the function below with logic to calculate the numeric risk score for the given input.

Be sure to provide meaningful documentation and at least two test cases in your documentation.  Also make sure your code satisfies the test cases provided in the assert statements.

````python
def cha2ds2_vasc(age, sex, chf, hypertension, stroke, vascular, diabetes):
    ### YOUR SOLUTION HERE
    """ (int, str, bool, bool, bool, bool, bool) -> int
    This function takes in patient data about age, sex, and history for CHF, hypertension, stroke, vasular disease, and diabetes 
    to return a score for Atrial Fibrillation Stroke Risk.
    The score is an integer.
    
    >>> cha2ds2_vasc(30, 'Female', True, False, True, True, True)
    6
    
    >>> cha2ds2_vasc(88, 'Male', False, False, True, False, True)
    5
    """
    if age < 65:
        Age = 0
    elif age <= 74:
        Age = 1
    else:
        Age = 2
        
    if sex == 'Male':
        Sex = 0
    else:
        Sex = 1
        
    if chf == True:
        CHF = 1
    else:
        CHF = 0
    
    if hypertension == True:
        Hypertension = 1
    else:
        Hypertension = 0
    
    if stroke == True:
        Stroke = 2
    else:
        Stroke = 0
        
    if vascular == True:
        Vascular = 1
    else:
        Vascular = 0
    
    if diabetes == True:
        Diabetes = 1
    else:
        Diabetes = 0
    
    score = Age + Sex + CHF + Hypertension + Stroke + Vascular + Diabetes
    return score
````

## Part 2: HEART Score

The HEART score is a predictor for major cardiac events.  It requires 5 high-level inputs:
* History (Slightly / Moderately / Highly suspicious)
* EKG (Normal / Non-specific repolarization disturbance / Significant ST deviation)
* Age (Number)
* Risk Factors (Number of risk factors)
* Initial Troponin (Number of times the normal limit)

Fill out the function below with logic to calculate the numeric risk score for the given input.

Be sure to provide meaningful documentation and at least two test cases in your documentation. Also make sure your code satisfies the test cases provided in the assert statements.

````python
def heart(history, ekg, age, risks, troponin):
    ### YOUR SOLUTION HERE
    """ (string, string, int, int, float) -> int
    This function takes in the history, EKG, age, risk factors, and initial troponin amount to return the HEART score.
    The score is an integer.
    
    >>> heart('Highly suspicious', 'Normal', 35, 2, 1.1)
    4
    
    >>> heart('Moderately suspicious', 'Significant ST deviation', 83, 3, 1.5)
    8
    """
    if history == 'Slightly suspicious':
        Hist = 0
    elif history == 'Moderately suspicious':
        Hist = 1
    else:
        Hist = 2
    
    if ekg == 'Normal':
        EKG = 0
    elif ekg == 'Non-specific repolarization':
        EKG = 1
    else:
        EKG = 2
    
    if age < 45:
        Age = 0
    elif age <= 64:
        Age = 1
    else:
        Age = 2
    
    if risks <= 0:
        Risks = 0
    elif risks <= 2:
        Risks = 1
    else:
        Risks = 2
    
    if troponin <= 1.0:
        Troponin = 0
    elif troponin <= 3.0:
        Troponin = 1
    else:
        Troponin = 2
    
    score = Hist + EKG + Age + Risks + Troponin
    
    return score
````

## Part 3: Framingham Risk Score for Hard Coronary Heart Disease

The Framingham Risk Score for Hard Coronary Heart Disease is intended for non diabetic patients age 30-79 only.  So, if the patient's age is < 30 or > 79, your function should return `-1` rather than a specific risk score.

The Framingham Risk Score takes 7 inputs:
* Age (Number)
* Sex (Male / Female)
* Smoker (True / False)
* Total cholesterol (Number)
* HDL cholesterol (Number)
* Systolic BP (Number)
* Blood pressure being treated with medicines (No / Yes)

You'll not that rather than being a basic parametric equation, this is a regression function defined by coefficients.  It also requires you take the natural logarithm (`ln`) of many of the parameters.  To help you out, here's an example of how to interpret the formula provided in the website's **Evidence** tab.

Take special note of the footnotes in the logic:
* *Yes=1, No=0 (for Treated for blood pressure and Smoker)
* ** Men: if age >70, use ln(70) x Smoker. Women: if age >78, use ln(78) x Smoker.


---

These segments of the equation and the coefficient table...

> $ L_{Men} = \beta \times \ln(Age) + \beta \times \ln(cholesterol) ... $
>
> | Variable &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; | Men | Women
> | :---------------------------- | :--: | :--: |
> | $ \ln(Age) $ | 52.00961 | 31.764001
> | $ \ln(Cholesterol) $ | 20.014077 | 22.465206

Could be written in Python as follows:

```python
import math

L = 0
if sex == "Male":
    L += 52.00961 * math.ln(age)
    L += 20.014077 * math.ln(cholesterol)
else:
    L += 31.764001 * math.ln(age)
    L += 22.465206 * math.ln(cholesterol)
```

**ROUND YOUR FINAL RESULT TO 4 DECIMAL PLACES**

````python
import math
def framingham(age, sex, smoker, cholesterol, hdl, systolic, bp_treated):
    ### YOUR SOLUTION HERE
    """ (int, str, bool, int, int, int, int, bool) -> float
    This function returns the Framingham Risk Score with the given inputs of age, sex, smoker, cholestrerol, hdl cholesterol, systolic BP, and BP treatment.
    This function rounds to 4 decimal places.
    """
    if age < 30 or age > 79:
        score = -1
    else:
        L = 0
        if sex == "Male":
            L += 52.00961 * math.log(age)
            L += 20.014077 * math.log(cholesterol)
            L += -0.905964 * math.log(hdl)
            L += 1.305784 * math.log(systolic)
            L += -4.605038 * math.log(age) * math.log(cholesterol)
            L += -2.93323 * math.log(age) * math.log(age)
            L += -172.300168
            if bp_treated == True:
                L += 0.241549
            if smoker == True:
                L += 12.096316 
                if age > 70:
                    L += -2.84367 * math.log(70)
                else:
                    L += -2.84367 * math.log(age)
        
            score = 1 - 0.9402**math.exp(L)  
        
        else:
            L += 31.764001 * math.log(age)
            L += 22.465206 * math.log(cholesterol)
            L += -1.187731 * math.log(hdl)
            L += 2.552905 * math.log(systolic)
            L += -5.060998 * math.log(age) * math.log(cholesterol)
            L += -146.5933061
            if bp_treated == True:
                L += 0.420251
            if smoker == True:
                L += 13.07543
                if age > 78:
                    L += -2.996945 * math.log(78)
                else:
                    L += -2.996945 * math.log(age)
            
            score = 1 - 0.98767**math.exp(L)
    
    return round(score, 4)
````

## Part 4. Putting it all together

Now that we have our three scores, we need to put them together into an overall composite risk score for a whole group of patients.  Those patients are in a CSV file on the server called `/data/midterm_patients.csv`.  You can open this file in Jupyter by browsing to your Home icon -> from_instructor -> data. First, you'll notice that some of the column names are different between the data in the input file and the values that are expected by your functions above.  For example: "M" from the file needs to be turned into "Male" and "Yes" in the file needs to be turned into "True".  You will need to do conversions for all of the fields listed below:

After calculating these three risk scores, use the rules below to determine who is at highest risk.  To be classified as "High Risk" a patient must meet all three criteria below:
1. CHA2DS2_VASc >= 2
2. HEART >= 4
3. Framingham >= 3%

Your output for this function needs to be a list where each item in this contains `[patient, CHA2DS2_VASc, HEART, Framingham, High Risk]`

````python
parameter_map = {'M': 'Male',
                'F': 'Female',
                'No': False,
                'Yes': True,
                }
````

````python
def test_patients(filename):
    ### YOUR SOLUTION HERE
    """ (str) -> list
    This function takes patient data in CSV form.
    It returns the patient ID, CHA2DS2_VASc score, HEART score, Framingham score, and High Risk status.
    """
    import csv
    patient_data = []
    with open(filename) as f:
        csv_reader = csv.reader(f, delimiter=',')
        next(csv_reader)
        for row in csv_reader:
            #if row[0] != 'E57853':
                #continue
            #cha2ds2_vasc(age, sex, chf, hypertension, stroke, vascular, diabetes)
            cha2ds2_score = cha2ds2_vasc(int(row[1]), parameter_map[row[2]], parameter_map[row[3]], parameter_map[row[4]], parameter_map[row[5]], parameter_map[row[6]], parameter_map[row[7]])
            #heart(history, ekg, age, risks, troponin)
            heart_score = heart(row[8], row[9], int(row[1]), int(row[10]), float(row[11]))
            #framingham(age, sex, smoker, cholesterol, hdl, systolic, bp_treated)
            framingham_score = framingham(int(row[1]), parameter_map[row[2]], parameter_map[row[12]], int(row[13]), int(row[14]), int(row[15]), parameter_map[row[16]])
            
            high_risk = False
            if cha2ds2_score >= 2:
                if heart_score >= 4:
                    if framingham_score >= 0.03:
                        high_risk = True
    
            array =[row[0], round(float(cha2ds2_score), 1), float(heart_score), float(framingham_score), high_risk]
        
            patient_data.append(array)
    
    return patient_data
````
